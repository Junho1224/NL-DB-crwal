```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import requests
import time
import os

def scroll_down(driver):
    """ 페이지 끝까지 스크롤하기 """
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)  # 페이지가 로드될 시간을 기다립니다.
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

# CSV 파일 읽기 및 매물 이름 중복 제거
df = pd.read_csv('./dataNop/apartment.csv', encoding='utf-8')
unique_names = df['atclNm'].drop_duplicates().tolist()

# 이미지를 저장할 폴더 생성
output_folder = 'downloaded_images'
os.makedirs(output_folder, exist_ok=True)

# WebDriver 설정 및 URL 열기
driver = webdriver.Chrome(service=Service())
try:
    driver.get('https://www.dabangapp.com/')
    time.sleep(5)  # 웹사이트가 완전히 로드될 때까지 대기

    for name in unique_names:
        try:
            # 검색 입력 필드
            search_input = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, '#search-input'))
            )
            search_input.clear()  # 기존 입력값 제거
            search_input.send_keys(name)
            time.sleep(3)

            # 드롭다운 리스트에서 원하는 항목 클릭
            try:
                search_result_item = WebDriverWait(driver, 10).until(
                    EC.element_to_be_clickable((By.CSS_SELECTOR, '#search-complex-list > div > div > button > div > p'))
                )
                driver.execute_script("arguments[0].click();", search_result_item)
                time.sleep(5)
            except Exception as e:
                print(f"드롭다운 선택 중 오류 발생: {e}")
                continue  # 다음 항목으로 넘어감

            # 매물 선택
            try:
                prop = WebDriverWait(driver, 10).until(
                    EC.element_to_be_clickable((By.CSS_SELECTOR, '#complex-list > div > div > a > h1'))
                )
                driver.execute_script("arguments[0].click();", prop)
                time.sleep(5)
            except Exception as e:
                print(f"매물 선택 중 오류 발생: {e}")
                continue  # 다음 항목으로 넘어감

            # 새 탭으로 전환
            original_window = driver.current_window_handle
            windows = driver.window_handles
            for window in windows:
                if window != original_window:
                    driver.switch_to.window(window)
                    break

            # 페이지 전체 스크롤
            scroll_down(driver)

            # 이미지 요소 찾기
            image_url = driver.execute_script("""
                var img = document.querySelector('img[alt="평면도 이미지"]');
                if (img) {
                    return img.src;
                } else {
                    var imgs = document.querySelectorAll('img');
                    for (var i = 0; i < imgs.length; i++) {
                        if (imgs[i].src.includes('d2o59jgeq8ig2.cloudfront.net')) {
                            return imgs[i].src;
                        }
                    }
                }
                return null;
            """)

            # 이미지가 있을 경우에만 다운로드
            if image_url:
                print(f"이미지 URL: {image_url}")
                try:
                    # 이미지 다운로드
                    image_data = requests.get(image_url).content
                    # 매물 이름을 파일명에 포함
                    image_filename = f'{name}.jpg'.replace(' ', '_')
                    image_filepath = os.path.join(output_folder, image_filename)
                    with open(image_filepath, 'wb') as file:
                        file.write(image_data)
                    print(f"이미지가 '{image_filename}' 파일로 저장되었습니다.")
                except Exception as e:
                    print(f"이미지 다운로드 중 오류 발생: {e}")
            else:
                print(f"{name}에 대한 이미지를 찾지 못했습니다.")

            # 새로고침 후 원래 창으로 전환
            driver.close()
            driver.switch_to.window(original_window)
            driver.refresh()
            time.sleep(3)

        except Exception as e:
            print(f"{name} 처리 중 오류 발생: {e}")
            # 새로고침 후 원래 창으로 전환
            driver.close()
            driver.switch_to.window(original_window)
            driver.refresh()
            time.sleep(3)

finally:
    # 드라이버 종료
    driver.quit()
```

```python
import os
import time
import urllib.request
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service

# 설정
URL = "https://www.google.co.kr/imghp?hl=ko&tab=wi&ogbl"
keyword = "한신아파트"

# 저장 폴더 생성
if not os.path.isdir(keyword):
    os.makedirs(keyword)

# Chrome 드라이버 설정
options = webdriver.ChromeOptions()
# options.add_argument("--headless")  # UI 없이 실행 (필요에 따라 주석 해제)
options.add_argument("--no-sandbox")
driver = webdriver.Chrome(service=Service(), options=options)
driver.set_window_size(1280, 1024)
driver.get(URL)

# 검색 수행
googleSearchElement = driver.find_element(By.NAME, "q")
googleSearchElement.send_keys(keyword)
googleSearchElement.send_keys(Keys.RETURN)

# 페이지 로딩 대기
time.sleep(3)

# 이미지 요소 찾기
# Google 이미지의 작은 썸네일을 클릭하여 큰 이미지를 로드하는 방식
thumbnails = driver.find_elements(By.CSS_SELECTOR, "img.rg_i")

print("Number of images (thumbnails) : {}".format(len(thumbnails)))

# 이미지 다운로드
for i, thumbnail in enumerate(thumbnails):
    try:
        # 썸네일 클릭하여 큰 이미지 로드
        thumbnail.click()
        time.sleep(1)
        
        # 큰 이미지 요소 찾기
        large_image = driver.find_element(By.CSS_SELECTOR, "img.n3VNCb")
        src = large_image.get_attribute("src")
        
        # 이미지 다운로드
        if src.startswith('http'):
            urllib.request.urlretrieve(src, os.path.join(keyword, "{}-{}.jpg".format(keyword, i)))
            print(f"Downloaded image {i}")
        else:
            print(f"Skipped image {i} due to invalid src")
        
    except Exception as e:
        print(f"Error downloading image {i}: {e}")
        continue

# 드라이버 종료
driver.quit()
```
